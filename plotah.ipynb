```python
from matplotlib import pyplot as plt
import seaborn as sns

def _plot_series(df):
    xs = df['Timestamp']
    color_map = {
        #'probeC1': 'blue',
        #'probeC2': 'red',
        #'probeC3': 'green',
        #'probeC4': 'orange'

        '61084071': 'blue',
        '61834993': 'red',
        'PSIMCOSTA01': 'green',
        '61737310': 'orange'
    }

    ys_list = [
        #('probeC1', df['probeC1']),
        #('probeC2', df['probeC2']),
        #('probeC3', df['probeC3']),
        #('probeC4', df['probeC4'])

        ('61834993', df['probeC1']),
        ('PSIMCOSTA01', df['probeC2']),
        #('61084071', df['probeC3']),
        ('61737310', df['probeC4'])
    ]

    for label, ys in ys_list:
        plt.plot(xs, ys, label=label, color=color_map[label])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df.sort_values('Timestamp', ascending=True)
# Removed the extra empty string argument
_plot_series(df_sorted)
sns.despine(fig=fig, ax=ax)
plt.xlabel('Timestamp')
plt.ylabel('probeC')
plt.legend()
plt.show()
```
